```python
 import numpy as np
import pandas as pd
import time
```

```python
df = pd.read_csv("/content/Social_Network_Ads.csv")
```

```python
df.head()
```

```python
df = df[['Age','EstimatedSalary','Purchased']]
```

```python
df.head()
```

```python
X = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

```python
X,y
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
X_scaled.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
X_train.shape
```

```python
X_train_scale = scaler.fit_transform(X_train)
```

```python
X_train_scale
```

```python
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(10,activation='relu',input_dim=2))
model.add(Dense(10,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',metrics=['accuracy'])
start = time.time()
history = model.fit(X_train_scale,y_train,epochs=10,batch_size=320)  #batch_size = 320 means Batch GD(vanilla GD)
print(time.time()-start)
```

```python
model.compile(loss='binary_crossentropy',metrics=['accuracy'])
start = time.time()
history = model.fit(X_train_scale,y_train,epochs=10,batch_size=1)  #batch_size = 1 means SGD
print(time.time()-start)
```

```python
#train time is faster in Batch GD but the convergence time in SGD is fast and as we can see we get the high accuracy in SGD i.e 0.9102
```
